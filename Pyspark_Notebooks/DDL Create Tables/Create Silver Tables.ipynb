# Silver Table

```python
%sql
CREATE TABLE IF NOT EXISTS silver.clean.acra (
    uen STRING,
    company_name STRING,
    entity_type_description STRING,
    entity_status_description STRING,
    registration_incorporation_date STRING,
    industry_code STRING,
    industry_description STRING,
    secondary_ssic_code STRING,
    secondary_ssic_description STRING,
    no_of_officers STRING,
    address STRING,
    founding_year INT,
    source_data STRING,
    created_at TIMESTAMP,
    updated_at TIMESTAMP
)
USING DELTA
LOCATION 'abfss://silver@singaporecomadls.dfs.core.windows.net/clean/acra';
```

```python
%sql
CREATE TABLE IF NOT EXISTS silver.clean.recordowl (
    uen_match STRING,
    owl_company_name STRING,
    recordowl_website STRING,
    company_website STRING,
    linkedin_url STRING,
    facebook_url STRING,
    instagram_url STRING,
    phone_number STRING,
    company_description STRING,
    owl_ssic_code STRING,
    owl_industry STRING,
    owl_secondary_ssic STRING,
    owl_secondary_industry STRING,
    source_data STRING,
    created_at TIMESTAMP,
    updated_at TIMESTAMP
)
USING DELTA
LOCATION 'abfss://silver@singaporecomadls.dfs.core.windows.net/clean/recordowl';
```

```python
%sql
CREATE TABLE IF NOT EXISTS silver.clean.scrapped_wesbites (
    uen_match STRING,
    recordowl_website STRING,
    scraped_company_name STRING,
    scraped_linkedin STRING,
    scraped_facebook STRING,
    scraped_instagram STRING,
    scraped_email STRING,
    scraped_phone STRING,
    scraped_keywords STRING,
    source_data STRING,
    created_at TIMESTAMP,
    updated_at TIMESTAMP
)
USING DELTA
LOCATION 'abfss://silver@singaporecomadls.dfs.core.windows.net/clean/scrape_websites';
```

```python
%sql
CREATE TABLE IF NOT EXISTS silver.clean.stocks (
    stock_symbol STRING,
    stock_company_name STRING,
    market_cap DOUBLE,
    revenue DOUBLE,
    stock_price DOUBLE,
    percent_change DOUBLE,
    source_data STRING,
    created_at TIMESTAMP,
    updated_at TIMESTAMP
)
USING DELTA
LOCATION 'abfss://silver@singaporecomadls.dfs.core.windows.net/clean/stocks';
```
